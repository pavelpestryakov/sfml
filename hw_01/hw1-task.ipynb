```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000

#TODO
def limit_case(limit):
    if  limit <= 10000.0:   #1
        return 'A'
    elif 10000.0 < limit <= 100000.0:   #2
        return 'B'
    elif 100000.0 < limit <= 200000.0:   #3
        return 'C'
    elif 200000.0 < limit <= 400000.0:   #4
        return'D'
    elif 400000.0 < limit <= 700000.0:   #5
        return 'E'
    elif limit > 700000.0:   #5
        return 'F'       


    # df['LIMIT_LEVEL'] = 
```

```python
# проверяем нашу функцию
for i in df['LIMIT_BAL'].head():
    print(limit_case(i))
```

```python
# создаем новую колонку 
df['LIMIT_LEVEL'] = df['LIMIT_BAL'].apply(limit_case)
```

```python
# проверяем
df[['LIMIT_LEVEL', 'ID']].groupby('LIMIT_LEVEL').count()
```
